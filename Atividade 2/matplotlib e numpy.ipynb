# Gr√°fico gaussianas

```python
'''
Plota 3 gr√°ficos de gaussiana com diferentes larguras e centros.

O linspace gera valores em um vetor para o eixo x, que ser√£o usados
para gerar valores de y que seguem a equa√ß√£o da gaussiana.
'''
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.linspace(0, 10, 100)
sigma1, sigma2, sigma3 = [0.1, 0.5, 0.9]
media1, media2, media3 = [1, 3, 5]
y1 = (1 / (2*np.pi*(sigma1)**2)**0.5)*np.exp(-0.5*((x - media1)/sigma1)**2)
y2 = (1 / (2*np.pi*(sigma2)**2)**0.5)*np.exp(-0.5*((x - media2)/sigma2)**2)
y3 = (1 / (2*np.pi*(sigma3)**2)**0.5)*np.exp(-0.5*((x - media3)/sigma3)**2)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.grid(True)
```

```python
'''
Gerador de valores

Este c√≥digo usa o random do numpy para gerar n√∫meros entre 0 e 10.
O loadtxt salva os termos gerados em um arquivo txt.
'''
import numpy as np
gerador = np.random.uniform(0, 10, size=(5000, 2))
np.savetxt('gauss.txt', gerador)
```

```python
'''
L√™ os dados salvos no arquivo gerado pelo c√≥digo anterior, e plota o
gr√°fico usando matplotlib.

Foi usado o plt.scatter pois o plot gerava um gr√°fico com linhas
sem sentido.
'''
import matplotlib.pyplot as plt
import numpy as np
valores = np.loadtxt('gauss.txt')
eixo_x = valores[:, 0]
sigma1, sigma2, sigma3 = [0.1, 0.5, 0.9]
media1, media2, media3 = [1, 3, 5]
y1 = (1 / (2*np.pi*(sigma1)**2)**0.5)*np.exp(-0.5*((eixo_x - media1)/sigma1)**2)
y2 = (1 / (2*np.pi*(sigma2)**2)**0.5)*np.exp(-0.5*((eixo_x - media2)/sigma2)**2)
y3 = (1 / (2*np.pi*(sigma3)**2)**0.5)*np.exp(-0.5*((eixo_x - media3)/sigma3)**2)
plt.scatter(eixo_x, y1)
plt.scatter(eixo_x, y2)
plt.scatter(eixo_x, y3)
plt.grid(True)
```

# Gr√°fico da fun√ß√£o exponencial

```python
'''
O c√≥digo gera um vetor com valores para o eixo x, e os utiliza para
gerar os valores das exponenciais com diferentes par√¢metros.

A fun√ß√£o plt.plot plota o gr√°fico com os par√¢metros do eixo x e y.
A fun√ß√£o plt.legend gera as legendas escritas no plot.
'''
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.linspace(-4, 4, 500)
f1 = np.exp(-10*x)
f2 = np.exp(-1*x)
f3 = np.exp(2*x)
f4 = np.exp(10*x)
plt.ylim(0, 10)
plt.plot(x, f1, label=-10)
plt.plot(x, f2, label=-1)
plt.plot(x, f3, label=2)
plt.plot(x, f4, label=10)
plt.legend()
plt.grid(True)
```

```python
'''
Gerador de valores

Este c√≥digo usa o random do numpy para gerar n√∫meros entre -4 e 4.
O loadtxt salva os termos gerados em um arquivo txt.
'''
import numpy as np
gerador = np.random.uniform(-4, 4, size=(800, 2))
np.savetxt('exponencial.txt', gerador)
```

```python
'''
L√™ os dados salvos no arquivo gerado pelo c√≥digo anterior, e plota o
gr√°fico usando matplotlib.

Foi usado o plt.scatter pois o plot gerava um gr√°fico com linhas
sem sentido.
'''
import matplotlib.pyplot as plt
import numpy as np
valores = np.loadtxt('exponencial.txt')
x = valores[:, 0]
f1 = np.exp(-10*x)
f2 = np.exp(-1*x)
f3 = np.exp(2*x)
f4 = np.exp(10*x)
plt.ylim(0, 10)
plt.scatter(x, f1, label=-10)
plt.scatter(x, f2, label=-1)
plt.scatter(x, f3, label=2)
plt.scatter(x, f4, label=10)
plt.legend()
plt.grid(True)
```

# Gr√°fico par√°bola

```python
'''
Gr√°fico da par√°bola

Este c√≥digo l√™ 3 par√¢metros (a, b, c) e plota o gr√°fico da par√°bola seguindo a
equa√ß√£o do segundo grau. as ra√≠zes s√£o calculadas utilizando bhaskara e o
comando plt.scatter gera pontos vermelhos onde o gr√°fico corta o eixo x.
'''
import numpy as np
import matplotlib.pyplot as plt
a1, b1, c1 = [1, 5, 4]
a2, b2, c2 = [1, 4, 4]
a3, b3, c3 = [1, 2, 3]
delta1 = (b1**2 - 4*a1*c1)
delta2 = (b2**2 - 4*a2*c2)
delta3 = (b3**2 - 4*a3*c3)
x11 = (-b1 + ((delta1)**(1/2)))/(2*a1)
x12 = (-b1 - (delta1)**(1/2))/(2*a1)
x21 = (-b2 + ((delta2)**(1/2)))/(2*a2)
x22 = (-b2 + ((delta2)**(1/2)))/(2*a2)
x31 = (-b3 + ((delta3)**(1/2)))/(2*a3)
x32 = (-b3 + ((delta3)**(1/2)))/(2*a3)
x1 = [x11, x12]
x2 = [x21, x22]
x3 = [x31, x32]
x = np.linspace(-10, 10, 100)
y1 = a1*(x)**2 + b1*x + c1
y2 = a2*(x)**2 + b2*x + c2
y3 = a3*(x)**2 + b3*x + c3
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.scatter(x1, [0, 0], c='red')
plt.scatter(x2, [0, 0], c='red')
plt.scatter(x3, [0, 0], c='red')
plt.grid(True)
```

```python
'''
Gerador de valores

Este c√≥digo usa o random do numpy para gerar n√∫meros entre -10 e 10.
O loadtxt salva os termos gerados em um arquivo txt.
'''
import numpy as np
gerador = np.random.uniform(-10, 10, size=(800, 2))
np.savetxt('parabola.txt', gerador)
```

```python
'''
L√™ os dados salvos no arquivo gerado pelo c√≥digo anterior, e plota o
gr√°fico usando matplotlib.

Foi usado o plt.scatter pois o plot gerava um gr√°fico com linhas
sem sentido.
'''
import matplotlib.pyplot as plt
import numpy as np
valores = np.loadtxt('parabola.txt')
x = valores[:, 0]
import numpy as np
import matplotlib.pyplot as plt
a1, b1, c1 = [1, 5, 4]
a2, b2, c2 = [1, 4, 4]
a3, b3, c3 = [1, 2, 3]
delta1 = (b1**2 - 4*a1*c1)
delta2 = (b2**2 - 4*a2*c2)
delta3 = (b3**2 - 4*a3*c3)
x11 = (-b1 + ((delta1)**(1/2)))/(2*a1)
x12 = (-b1 - (delta1)**(1/2))/(2*a1)
x21 = (-b2 + ((delta2)**(1/2)))/(2*a2)
x22 = (-b2 + ((delta2)**(1/2)))/(2*a2)
x31 = (-b3 + ((delta3)**(1/2)))/(2*a3)
x32 = (-b3 + ((delta3)**(1/2)))/(2*a3)
x1 = [x11, x12]
x2 = [x21, x22]
x3 = [x31, x32]
y1 = a1*(x)**2 + b1*x + c1
y2 = a2*(x)**2 + b2*x + c2
y3 = a3*(x)**2 + b3*x + c3
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.scatter(x, y3)
plt.scatter(x1, [0, 0], c='red')
plt.scatter(x2, [0, 0], c='red')
plt.scatter(x3, [0, 0], c='red')
plt.grid(True)
```

# Gr√°fico Deltoid

```python
'''
Gr√°fico Deltoid

Gera n√∫meros igualmente espa√ßados para o par√¢metro theta, que
ser√£o usados para determinar os valores de x e y.
'''
import numpy as np
import matplotlib.pyplot as plt
theta = np.linspace(0, 2*np.pi, 500)
x = 2*np.cos(theta)+np.cos(2*theta)
y = 2*np.sin(theta)+np.sin(2*theta)
plt.plot(x, y)
plt.grid(True)
```

# Gr√°fico espiral de Galileu

```python
'''
Gr√°fico da espiral de Galileu

Este c√≥digo gera 100 valores entre 0 e 2pi em um array, a coordenada
radial √© determinada pela fun√ß√£o theta**2.

A fun√ß√£o plt.axes passa de coordenadas cartesianas para polares
'''
import numpy as np
import matplotlib.pyplot as plt
import math
theta = np.linspace(0, 4*np.pi, 100)
r = theta**2
plt.axes(polar=True)
plt.plot(theta, r)
```

# Fun√ß√£o de Fey

```python
'''
Gr√°fico da fun√ß√£o de Fey

Este c√≥digo gera 100 valores entre 0 e 2pi em um array, que ser√£o
os valores da coordenada angular

A fun√ß√£o plt.axes passa de coordenadas cartesianas para polares
'''
import numpy as np
import matplotlib.pyplot as plt
import math
theta = np.linspace(0, 6*np.pi, 500)
r = np.exp(np.cos(theta)) - 2*np.cos(4*theta) + (np.sin(theta/12)**5)
plt.axes(polar=True)
plt.plot(theta, r)
```
