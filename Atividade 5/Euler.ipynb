```python
import numpy as np
import matplotlib.pyplot as plt

v0 = 0
v_t = [0]
tempo = np.linspace(0.1, 100, 1000)

for i in range(1, 1000):
  Vn = v0 + (-9.8 - 0.1*v0)*0.1
  v0 = Vn
  v_t.append(Vn)

plt.plot(tempo, v_t)
plt.show()
```

```python
''' 
RK4
'''
import numpy as np
import matplotlib.pyplot as plt

y0 = v0 = 0
tempo = np.linspace(0, 100, 1000)
h = tempo[1] - tempo[0]

def f(v0):
  f = -9.8 - 0.1*v0
  return f
  
v_rk4 = [0]
for i in range(1, len(tempo)):
  k1 = f(y0)
  k2 = f(y0 + (h/2)*k1)
  k3 = f(y0 + (h/2)*k2)
  k4 = f(y0 + (h*k3))
  yn = y0 + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
  v_rk4.append(yn)
  y0 = yn

v_euler = [0]
for i in range(1, len(tempo)):
  Vn = v0 + (-9.8 - 0.1*v0)*h
  v0 = Vn
  v_euler.append(Vn)

plt.plot(tempo, v_rk4, label='RK4')
plt.plot(tempo, v_euler, label='Euler')
plt.legend()
plt.grid()
plt.show()
```
