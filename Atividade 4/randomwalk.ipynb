# Passeio Aleatorio

```python
'''
Gera números pseudoaleatórios para usar na probabilidade, dessa maneira é possível definir 
qual a direção da posição baseada no número gerado.
'''
import numpy as np
import matplotlib.pyplot as plt

N = 1000
n = np.linspace(1, N, N)
rng = np.random.default_rng(123)
def passeio_aleatorio(N, p):
  x = 0
  posicao = [0]
  for i in range(1, N):
    sorteio = rng.random()
    if sorteio < p:
      x += 1
      posicao.append(x)
    else:
      posicao.append(x)
      x -= 1
  return posicao

posicao1 = passeio_aleatorio(1000, 0.5)
posicao2 = passeio_aleatorio(1000, 0.8)
plt.plot(n, posicao1, 'o-', label='p = 0.5', ms=1)
plt.plot(n, posicao2, 'o-', label='p = 0.8', ms=1)
plt.legend()
plt.grid()
```

```python
'''
Faz um loop for para plotar cada gráfico 100 vezes, isso para cada valor de p = 0.5 e 0.8
'''
for i in range(1, 100):
    y = passeio_aleatorio(1000, 0.5)
    plt.plot(n, y, c='r', alpha=0.1)
for i in range(1, 100):
    y = passeio_aleatorio(1000, 0.6)
    plt.plot(n, y, c='b', alpha=0.1)
```

```python
def media_dp(N, p, num_passeios):
  media = []
  dp = []
  for i in range(1, N+1):
    posicao_final = []
    for _ in range(1, num_passeios):
        posicao = passeio_aleatorio(i,p)
        posicao_final.append(posicao[-1])
    media.append(np.mean(posicao_final))
    dp.append(np.std(posicao_final))
  return np.array(media), np.array(dp)
N=1000
num_passeios=100
n = np.linspace(1, N, N)
media, dp = media_dp(1000, 0.8, num_passeios)
plt.plot(n, media)
plt.plot(n, dp)
```
