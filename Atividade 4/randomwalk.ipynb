# Passeio Aleatorio

```python
'''
Cria um vetor com mil valores sorteados, sendo eles -1 ou 1. Caso o sorteio seja
1, o 'bêbado' vai para um lado, se for -1 vai para o outro.
A soma cumulativa faz o vetor com as posições em cada instante de tempo.
'''
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 1000, 1000)

sorteio = np.random.choice([-1, 1], [1000])
posicao1 = np.cumsum(sorteio, axis=0)

sorteio = np.random.choice([-1, 1], [1000], p=[0.4, 0.6])
posicao2 = np.cumsum(sorteio, axis=0)

posicao1[0] = posicao2[0] = 0
plt.plot(n, posicao1)
plt.plot(n, posicao2)
```

```python
'''
Cria uma matriz 1000 por 100 com valores pseudoaleatórios, podendo ser 1 ou -1, os quais
definirão se o 'bêbado' vai para a direita ou para a esquerda. a função cumsum faz a soma cumulativa
e defini o trajeto.
O parâmetro p define as probabilidades do bêbado ir para a direita ou esquerda.
'''
import numpy as np
import matplotlib.pyplot as plt

gerador = np.random.choice([-1, 1], [1000, 100], p=[0.5, 0.5])
posicao1 = np.cumsum(gerador, axis=0)
posicao1 = posicao1 - posicao1[0]

gerador = np.random.choice([-1, 1], [1000, 100], p=[0.4, 0.6])
posicao2 = np.cumsum(gerador, axis=0)
posicao2 = posicao2 - posicao2[0]

plt.plot(posicao1, c='r')
plt.plot(posicao2, c='b')
plt.show()
```

```python
'''
Calcula a média e o desvio padrão usando as funções np.mean e np.std. 
O parâmetro axis diz se o cálculo será feito nas linhas ou colunas.
No caso mostrado é feito nas linhas, uma vez que são as posições para um
mesmo instantede tempo.
'''
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 1000, 1000)

gerador = np.random.choice([-1, 1], [1000, 100], p=[0.5, 0.5])
posicao1 = np.cumsum(gerador, axis=0)
posicao1 = posicao1 - posicao1[0]

gerador = np.random.choice([-1, 1], [1000, 100], p=[0.4, 0.6])
posicao2 = np.cumsum(gerador, axis=0)
posicao2 = posicao2 - posicao2[0]

media1 = np.mean(posicao1, axis=1) # axis=1 se refere à média feita nas linhas
media2 = np.mean(posicao2, axis=1)

dp1 = np.std(posicao1, axis=1)
dp2 = np.std(posicao2, axis=1)

plt.plot(media1)
plt.plot(media2)
plt.errorbar(n, media1, yerr=dp1, alpha=0.1)
plt.errorbar(n, media2, yerr=dp2, alpha=0.1)
plt.show()
```
