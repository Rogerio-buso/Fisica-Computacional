# Mapa Log√≠stico

```python
''''
Faz uma evolu√ß√£o temporal usando o loop for da equa√ß√£o x=r*x0(1-x0). A partir 
disso a fun√ß√£o retorna uma lista com os valores em cada tempo.
'''

import matplotlib.pyplot as plt
import numpy as np
def mapa_logistico(r, x0, N):
  lista = [x0]
  for i in range (0, N):
    x = r*x0*(1 - x0)
    lista = lista + [x]
    x0 = x
  return lista

pontos = 200
n = np.arange(0, pontos + 1)
y1 = mapa_logistico(2, 0.3, pontos)
y2 = mapa_logistico(3.3, 0.3, pontos)
y3 = mapa_logistico(3.5, 0.3, pontos)
y4 = mapa_logistico(3.9, 0.3, pontos)
plt.plot(n, y1, 'o-')
plt.show()
plt.plot(n, y2, 'o-', ms=1)
plt.show()
plt.plot(n, y3, 'o-', ms=1)
plt.show()
plt.plot(n, y4, 'o-', ms=1)
plt.show()
```

```python
'''
Primeiro foi criado um vetor rs que gera 50 valores de 1 a 4.
O vetor xs gera 50 vetores bidimensionais com valores aleat√≥rios.
O for com √≠ndice n faz a evolu√ß√£o temporal.
O for com √≠ndice 'j' e 'i' s√£o respons√°veis por localizar a posi√ß√£o dos elementos de vetores
que devem ser multiplicados na equa√ß√£o.
'''

import matplotlib.pyplot as plt
import numpy as np
nr = 50
nx = 50
rs = np.linspace(1, 4, nr)
xs = np.random.random([nr, nx])
for i in range(0, nr):
  for j in range(0, nx):
    for n in range(0, 100):
      xs[i,j] = rs[i]*xs[i,j]*(1-xs[i,j])
plt.plot(rs, xs, 'o', color='red', ms=1)
plt.show()
```

```python
'''
Primeiro foi criado uma matriz 50 x 100 com valores pseudoaleat√≥rios.
O vetor rs possui 50 valores entre 1 e 4.
Ap√≥s isso, o vetor √© transformado em uma lista, multiplicado pela vari√°vel x, transformado
novamente em um array, e por fim transpondo-o.
O for faz a evolu√ß√£o temporal com os valores a partir da 1000¬∞ tempo at√© o 2000¬∞ utilizando 
multiplica√ß√£o e soma de matrizes, o que deixa o c√≥digo mais eficiente.
'''

import matplotlib.pyplot as plt
import numpy as np

r = 50
x = 100
xs = np.random.random([r, x])
rs = np.linspace(1, 4, r)
rs = np.array([rs.tolist()]*x).T
for n in range(1000, 2000):
  xs = rs*xs*(1-xs)
plt.scatter(rs, xs, color='red', s=1)
plt.show()
```
