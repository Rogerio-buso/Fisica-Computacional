# Mapa Log√≠stico

```python
import matplotlib.pyplot as plt
import numpy as np
def mapa_logistico(r, x0, N):
  lista = [x0]
  for i in range (0, N):
    x = r*x0*(1 - x0)
    lista = lista + [x]
    x0 = x
  return lista

pontos = 90
n = np.arange(0, pontos + 1)
y1 = mapa_logistico(2, 0.3, pontos)
y2 = mapa_logistico(3.3, 0.3, pontos)
y3 = mapa_logistico(3.5, 0.3, pontos)
y4 = mapa_logistico(3.9, 0.3, pontos)
plt.plot(n, y1, 'o-')
plt.show()
plt.plot(n, y2, 'o-', ms=1)
plt.show()
plt.plot(n, y3, 'o-', ms=1)
plt.show()
plt.plot(n, y4, 'o-', ms=1)
plt.show()
```

```python
nr = 50
nx = 5
rs = np.linspace(0, 4, nr)
xs = np.random.random([nr, nx])
for i in range(0, nr):
  for j in range(0, nx):
    for n in range(0, 100):
      xs[i,j] = rs[i]*xs[i,j]*(1-xs[i,j])
plt.plot(rs, xs, 'o', color='red', ms=1)
%time
```

```python
r = 50
x = 100
xs = np.random.random([r, x])
rs = np.linspace(0, 4, r)
rs = np.array([rs.tolist()]*x).T
for n in range(1000, 2000):
  xs = rs*xs*(1-xs)
plt.scatter(rs, xs, color='red', s=1)
%time
```
