# Monte Carlo

```python
'''
Monte Carlo
Calcula o valor de pi através da propabilidade deum ponto aleatório cair
dentro de um quarto de circunferência de raio 1, sendo que a área total é um quadrado
de área 1.
'''
import numpy as np
def pi_monte_carlo(N):
  num = np.random.uniform(0, 1, size=(N, 2))
  x = num[:,0]
  y = num[:,1]
  j = 0
  for i in range(0, N):
    r = ((x[i])**2 + (y[i])**2)**0.5
    if r <= 1:
      j += 1
  return j, N
P = pi_monte_carlo(10000)[0]/pi_monte_carlo(10000)[1]
print(4*P)
```

```python
import matplotlib.pyplot as plt
import numpy as np
def pi_monte_carlo(N, retorna_pontos = False):
  num = np.random.uniform(0, 1, size=(N, 2))
  x = num[:,0]
  y = num[:,1]
  j = 0
  dentro = []
  fora = []
  for i in range(0, N):
    r = ((x[i])**2 + (y[i])**2)**0.5
    if r <= 1:
      j += 1
      dentro.append((x[i], y[i]))
    else:
      fora.append((x[i], y[i]))
  p = 4*(j/N)
  if retorna_pontos == True:
    return p, dentro, fora
  else:
    return p
pi, dentro, fora = pi_monte_carlo(10000, retorna_pontos=True)
x_dentro, y_dentro = zip(*dentro)
x_fora, y_fora = zip(*fora)
plt.scatter(x_dentro, y_dentro, color='blue', s=1)
plt.scatter(x_fora, y_fora, color='red', s=1)
plt.gca().set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pi_monte_carlo(N):
  num = np.random.uniform(0, 1, size=(N, 2))
  x = num[:,0]
  y = num[:,1]
  j = 0
  for i in range(0, N):
    r = ((x[i])**2 + (y[i])**2)**0.5
    if r <= 1:
      j += 1
  P = j/N
  pi = 4*P
  return pi

En = []
N = []
p = 10

for i in range(1, 7):
  N = N + [p]
  cont = 0
  for j in range(1, 10):
    Erro = ((abs(np.pi - pi_monte_carlo(p)))/np.pi)
    cont += Erro
  p *= 10
  media = cont/10
  En.append(media)
```

```python
plt.subplot(221)
plt.plot(N, En, 'o')

plt.subplot(222)
plt.loglog(N, En, 'o')

plt.subplot(223)
plt.semilogx(N, En, 'o')

plt.subplot(224)
plt.semilogy(N, En, 'o')

plt.tight_layout()
plt.show()
```

```python
x1 = np.log(N)
y1 = np.log(En)
param = np.polyfit(x1, y1, 1)
plt.loglog(N, En, 'o')
x_fit = np.linspace(10, 1000000, 100)
y_fit = np.polyval(param, x_fit)
plt.plot(x_fit, y_fit,'r--')
print(param)
```
