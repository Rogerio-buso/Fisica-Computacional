# Monte Carlo

```python
'''
Monte Carlo
Calcula o valor de pi através da propabilidade deum ponto aleatório cair
dentro de um quarto de circunferência de raio 1, sendo que a área total é um quadrado
de área 1
'''
import numpy as np
def pi_monte_carlo(N):
  num = np.random.uniform(0, 1, size=(N, 2))
  x = num[:,0]
  y = num[:,1]
  j = 0
  for i in range(0, N):
    r = ((x[i])**2 + (y[i])**2)**0.5
    if r <= 1:
      j += 1
  return j, N
P = pi_monte_carlo(10000)[0]/pi_monte_carlo(10000)[1]
print(4*P)
```

```python
import matplotlib.pyplot as plt
import numpy as np
def pi_monte_carlo(N, retorna_pontos = False):
  num = np.random.uniform(0, 1, size=(N, 2))
  x = num[:,0]
  y = num[:,1]
  j = 0
  dentro = []
  fora = []
  for i in range(0, N):
    r = ((x[i])**2 + (y[i])**2)**0.5
    if r <= 1:
      j += 1
      dentro.append((x[i], y[i]))
    else:
      fora.append((x[i], y[i]))
  p = 4*(j/N)
  if retorna_pontos == True:
    return p, dentro, fora
  else:
    return p
pi, dentro, fora = pi_monte_carlo(10000, retorna_pontos=True)
x_dentro, y_dentro = zip(*dentro)
x_fora, y_fora = zip(*fora)
plt.scatter(x_dentro, y_dentro, color='blue', s=1)
plt.scatter(x_fora, y_fora, color='red', s=1)
plt.show()
```

# Mapa logistico

```python
import matplotlib.pyplot as plt
import numpy as np
def mapa_logistico(r, x0, N):
  lista = [x0]
  for i in range (0, N):
    x = r*x0*(1 - x0)
    lista = lista + [x]
    x0 = x
  return lista
n = np.linspace(0, 1000, 1001)
y1 = mapa_logistico(2, 1/2, 1000)
y2 = mapa_logistico(3.3, 1/2, 1000)
y3 = mapa_logistico(3.5, 1/2, 1000)
y4 = mapa_logistico(3.9, 1/2, 1000)
plt.plot(n, y1)
plt.show()
plt.scatter(n, y2, s=1)
plt.show()
plt.scatter(n, y3, s=1)
plt.show()
plt.scatter(n, y4, s=1)
plt.show()
```
